```python
import pandas as pd
import numpy as np
```

```python
temp_df = pd.read_excel("../_data/RFC Tables/EL Computational Tools data model.xlsx")

temp_df = temp_df.rename(
    columns={
        "key": "Attribute",
        "description": "Description",
        "required": "Required",
        "requires": "Module",
        "concept source ontology": "Ontology",
        "valid values": "Valid Values",
        "type": "columnType",
        "note": "Notes",
    }, errors = 'ignore'
)

temp_df = temp_df.fillna("")
```

```python
# remove unneeded characters
temp_df = temp_df.replace('\n', '', regex = True)
```

```python
temp_df['Module'] = 'Tool'
```

```python
temp_df['UsedIn'] = 'ComputationalTools'
```

```python
template = {
    "Attribute": ["ComputationalTools"],
    "Description": ["Template used to curate computational tools used by researchers"],
    "Valid Values": [""],
    "DependsOn": ['Component' + ',' + ",".join(temp_df.Attribute)],
    "Properties": ["Template"],
    "Required": ["True"],
    "Parent": ["Template"],
    "DependsOn Component": [""],
    "Source": ["Sage Bionetworks"],
    "Validation Rules": [""],
    "Module": ["Template"],
    "Ontology": ["Sage Bionetworks"],
    "UsedIn": ["ComputationalTools"],
    "columnType": [""],
    "Notes": [""],
}
```

```python
temp_df = pd.concat([temp_df, pd.DataFrame(template)])
```

```python
temp_df = temp_df.set_index("Attribute")
temp_df
```

```python
temp_df
```

```python
# Load and join with exisiting data model
dm = pd.read_csv("../EL.data.model.csv", index_col='Attribute')
dm = dm.drop(columns=["OtherValue"], errors='ignore')
dm
```

```python
dm_final = pd.concat([dm, temp_df[~temp_df.index.isin(dm.index)]])
```

```python
if dm_final.index.is_unique:
    dm_final.to_csv('../EL.data.model.csv')
    print('Updated data model')
```
